```python
path = '/Users/lucialarraona/Desktop/projects_sem2/dtudeeplearning22/Files/iSarcasm/isarcasm_test.csv'
df = pd.read_csv(path)
#df = df.dropna()
#df = df.reset_index()
df = df.fillna('none')
df
```

```python
'''
Gets text content for tweet IDs
'''

# third-party: `pip install tweepy`
import tweepy

# api keys from twitter developers
consumer_key=''
consumer_secret=''
access_token=''
acess_token_sectret=''

# login to twitter 
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, acess_token_sectret)
auth.secure = True
api = tweepy.API(auth)
```

```python
def get_tweet_text(id):
    try:
        return api.get_status(id).text
    except:
        return "Tweet no longer exists"

df['tweet'] = df.tweet_id.apply(lambda x: get_tweet_text(x))
df
```
